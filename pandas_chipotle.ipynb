#data manipulation

```python
df.item_price.dtype
```

```python
df.item_price = df.item_price.apply(lambda x : float(x.replace('$', '')))
```

```python
df.head()
```

```python
df["total_price"] = df["item_price"]*df["quantity"]
```

```python
df.head()
```

#How much was the revenue for the period in the dataset?

```python
revenue = df["total_price"].sum()
print(revenue)
```
