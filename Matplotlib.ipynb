```python
x = np.array([1,2,3,4])
y = np.array([5,-1,-5,4])
plt.plot(x,y, color = 'green');
```

pyplot API & object-oriented API

```python
#pyplot API 
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color ="brown",label ="sin")
plt.plot(x,np.cos(x),color ="blue",label = "cos")
plt.xlabel("X")
plt.ylabel("F(x)")
plt.xlim([3,9])
plt.ylim([-2,2])
#plt.axis([xmin,xmax,ymin,ymax])
#plt.asix("tight")
#plt.asix("equal")
plt.legend()
plt.title("sin(x) and cos(x)");
```

```python
#OO API 
#1 prepare data
x = [1,4,5,6]
y = [11,44,23,12]

#2 set up plot
fig, ax = plt.subplots(figsize=(5,5))  #figure size = width and hight of plot

#3 plot of data 
ax.plot(x,y)
ax.set(title="x and y"
      ,xlabel="x"
      ,ylabel="y"
      )
;
```

types of matplotlib  plots

```python
#line
x=np.linspace(0,10,100)
#OO API plot a line chart

fig, ax = plt.subplots()
ax.plot(x,x**3);
```

```python
#scatter 
#pyplot API
plt.scatter(x,np.exp(x)); #y= e^x
```

```python
#scatter 
#OO API
#prepare data
rng= np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
sizes = 1000*rng.rand(100)
colors = 1000*rng.rand(100)
fig,ax = plt.subplots()
chart1= ax.scatter(x,y,s=sizes, c=colors, cmap = "viridis",alpha = 0.8);
fig.colorbar(chart1);
```

```python
#Bar (vertial, horizontal)
#prepare data
soft_drink = {"coca":10,"pesi":12,"mirinda":6}
fig, ax = plt.subplots()
ax.bar(soft_drink.keys(),soft_drink.values());
ax.set(title=("summer_drink")
      ,xlabel="beverge"
      ,ylabel="$");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink.keys()),list(soft_drink.values()));
```

```python
#histogram
np.random.seed(42)
student_height = np.random.normal(170,10,250)
```

```python
#pyplot API
```

```python
plt.hist(student_height,bins=30);
```

Subplots

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2
                                          ,ncols = 2
                                          ,figsize = (10,10))
#plot each ax
ax1.plot(x,x/2);
ax1.plot(x,x*6);
ax2.scatter(np.random.random(100),np.random.random(100));
ax3.barh(list(soft_drink.keys()),list(soft_drink.values()));
ax4.hist(student_height,bins=30);
```

```python
fig, ax = plt.subplots(nrows = 2
                        ,ncols = 2
                        ,figsize = (10,10))
#plot each ax
ax[0,0].plot(x,x/2);
ax[0,0].plot(x,x*6);
ax[0,1].scatter(np.random.random(100),np.random.random(100));
ax[1,0].barh(list(soft_drink.keys()),list(soft_drink.values()));
ax[1,1].hist(student_height,bins=30);

```
