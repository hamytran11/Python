```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
cities= pd.read_csv('california_cities.csv')
cities.head(5)
```

```python
#extract latd (vido) and logd(kinhdo)
```

```python
lat,lon = cities['latd'],cities['longd']
population, area = cities['population_total'],cities['area_total_km2']
#plt.style.use()
plt.figure(figsize=(10,8))
#plot using pyplot API
plt.scatter(lon,lat,c=np.log10(population),cmap='viridis',s=area, linewidths=0, alpha = 0.8)
plt.axis('equal')
plt.xlabel("longlatitude")
plt.ylabel("latitude")
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)
plt.title("California cities: Population and area distribution");
```
